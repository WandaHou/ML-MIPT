```python
import torch
from matplotlib import pyplot as plt
```

## Negativity plot

```python
mps_N_values = torch.load("mps_N_values.pt")
gpt_N_values = torch.load("gpt_N_values.pt")
ps_N_values = torch.load("ps_N_values.pt")

# Calculate means and standard deviations
x = torch.arange(4, 36, 2)  # x-axis values

# Calculate statistics for each dataset
mps_mean = mps_N_values.mean(dim=1)
mps_std = ((mps_N_values**2).mean(dim=1)-mps_mean**2)/mps_N_values.shape[-1]**0.5
gpt_mean = gpt_N_values.mean(dim=1)
gpt_std = ((gpt_N_values**2).mean(dim=1)-gpt_mean**2)/gpt_N_values.shape[-1]**0.5
ps_mean = ps_N_values.mean(dim=1)
ps_std = ((ps_N_values**2).mean(dim=1)-ps_mean**2)/ps_N_values.shape[-1]**0.5

# Create the plot
plt.figure(figsize=(10, 6))

# Plot each dataset with error bars
plt.errorbar(x, mps_mean, yerr=mps_std, label='MPS', marker='o', capsize=5)
plt.errorbar(x, gpt_mean, yerr=gpt_std, label='GPT', marker='s', capsize=5)
plt.errorbar(x, ps_mean, yerr=ps_std, label='Post-selection', marker='^', capsize=5)

# Customize the plot
plt.xlabel('System Size (N)', fontsize=12)
plt.ylabel('Negativity', fontsize=12)
plt.title('Negativity vs System Size', fontsize=14)
plt.grid(True, linestyle='--', alpha=0.7)
plt.legend(fontsize=10)
plt.axhline(0, color='k', linewidth=1.5, linestyle='--', alpha=0.8)

# Adjust layout and show
plt.tight_layout()
plt.show()
```
